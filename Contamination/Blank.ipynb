# Contamination chart

```python
# Axis setting
x = df['Sample Id']
y = df['Au']
```

```python
# Edit type of control, analyte and laboratory 
control = 'Blank'
analyte = 'Au'
lab = 'INTK'
```

```python
# Specification limit
limit = 0.09
```

```python
# Contamination plot
fig, ax = plt.subplots()

ax.set_title("Contamination Chart")

# Samples result lineplot
ax.plot(y, color='blue', linewidth=1, marker='o')

# Lineplot edgecolors for errors
for i in range(1, len(y)):
    if y[i] > limit:
        ax.scatter(x[i-1], y[i], edgecolors='red', color='white')
    else:
        ax.scatter(x[i-1], y[i], color='white')

# Control line for upper limit      
ax.hlines(y=limit, xmin=1, xmax=len(df), linewidth=1, color='red', linestyles='--', label='Upper Limit')
plt.legend()

# Edit Y range
plt.ylim([0.0, 0.1])

# Plot Y ticks
plt.yticks(np.arange(0.0, 0.15, 0.025))

# Red dots identification
colors = np.where(y > limit, 'r', 'w')

# Counting points over specification limit (red dots)
num_points_over_limit = np.sum(colors == 'r')

# x-axis rotation
plt.xticks(rotation=45)

# Grid plot
plt.grid()

# Axis titles
plt.xlabel('Sample ID')
plt.ylabel('Au g/t')

plt.show()
```

# Percentage of points over specification limit

```python
# Percentage of fails
fails = num_points_over_limit / len(x) * 100
```

```python
print("The percentage of fails of {} for {} in laboratory {} is {:.2f}".format(control,analyte,lab,fails))
```
